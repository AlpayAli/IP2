```python
import pdfplumber
import json
```

```python
file_path = "chatbot/data/BGC_texas.pdf"
with pdfplumber.open(file_path) as pdf:
    all_text = ""
    for page in pdf.pages:
        all_text += page.extract_text()

print(all_text)
```

```python
lines = all_text.split("\n")

sections = {}
current_section = None

for line in lines:
    line = line.strip()
    if "Hand Rankings" in line:
        current_section = "Handwaarden"
        sections[current_section] = []
    elif "Dealing Procedures" in line:
        current_section = "Speelrondes"
        sections[current_section] = []
    elif "How Winners are Determined" in line:
        current_section = "Showdown"
        sections[current_section] = []
    elif current_section: 
        sections[current_section].append(line)

for section, content in sections.items():
    print(f"== {section} ==")
    print("\n".join(content[:5]))  
```

```python
with open("chatbot/rules/structured_rules.json", "w") as f:
    json.dump(sections, f, indent=4)
```
